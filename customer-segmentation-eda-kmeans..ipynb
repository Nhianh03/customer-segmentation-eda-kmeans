## üìâ Elbow Curve 
#t·ªëi ∆∞u thu·∫≠t to√°n Kmeans

```python
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(new_data)
    wcss.append(kmeans.inertia_)
    plt.figure(figsize=(16,8))
plt.plot(range(1,11),wcss, 'bx-')
plt.title('The Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

* Trong bi·ªÉu ƒë·ªì, ƒëi·ªÉm khu·ª∑u c√≥ v·∫ª n·∫±m ·ªü kho·∫£ng 2 ho·∫∑c 3 c·ª•m. ƒêi·ªÅu n√†y cho th·∫•y r·∫±ng vi·ªác th√™m nhi·ªÅu c·ª•m h∆°n ƒëi·ªÉm n√†y s·∫Ω kh√¥ng mang l·∫°i nhi·ªÅu l·ª£i √≠ch v·ªÅ vi·ªác gi·∫£m WCSS.
* L·ª±a Ch·ªçn C·ª•m: l√† 2 ho·∫∑c 3 c·ª•m cho ph√¢n t√≠ch K-means

### Gi·∫£i Th√≠ch:

* N·∫øu ch·ªçn 3 c·ª•m, ƒëi·ªÅu n√†y cho ph√©p m·ªôt c√°i nh√¨n chi ti·∫øt h∆°n v·ªÅ d·ªØ li·ªáu, c√≥ th·ªÉ ti·∫øt l·ªô c√°c nh√≥m con trong c√°c c·ª•m ch√≠nh .
